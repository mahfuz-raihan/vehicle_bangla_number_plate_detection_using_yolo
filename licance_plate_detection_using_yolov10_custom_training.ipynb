````python
# Let's download the model from the model zoo

import os
import urllib.request

# create a directroy for the weights in the current working directory

weights_dir = os.path.join(os.getcwd(), "weights")
os.makedirs(weights_dir, exist_ok=True)

# URL of the weight file
urls = [
    'https://github.com/ultralytics/assets/releases/download/v8.2.0/yolov10n.pt',
    'https://github.com/ultralytics/assets/releases/download/v8.2.0/yolov10s.pt',
    'https://github.com/ultralytics/assets/releases/download/v8.2.0/yolov10m.pt',
    'https://github.com/ultralytics/assets/releases/download/v8.2.0/yolov10b.pt',
    'https://github.com/ultralytics/assets/releases/download/v8.2.0/yolov10l.pt',
    'https://github.com/ultralytics/assets/releases/download/v8.2.0/yolov10x.pt'
]

# Download each file
for url in urls:
    file_name = os.path.join(weights_dir, os.path.basename(url))
    urllib.request.urlretrieve(url, file_name)
    print(f"Downloaded {file_name}")
````
